```python
import pandas as pd

df = pd.read_csv('/CO2 Emissions_Canada.csv')

print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Engine Size(L)'], df['CO2 Emissions(g/km)'])
plt.xlabel('Tamaño del Motor (L)')
plt.ylabel('Emisiones de CO2 (g/km)')
plt.title('Relacion entre el tamaño del Motor y CO2 emitido')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x =df[['Engine Size(L)']]
y = df['CO2 Emissions(g/km)']

x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
plt.scatter(x_test, y_test, color='gray')
plt.plot(x_test, model.predict(x_test), color='red')
plt.xlabel('Tamaño del Motor (L)')
plt.ylabel('Emisiones de CO2 (g/km)')
plt.title(' Prediciones vs realidad')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

preds = model.predict(x_test)
print('R2 Score:', r2_score(y_test, preds))
print('MSE:',mean_squared_error(y_test,preds))

rmse = np.sqrt(949.9852533375358)
print('RMSE:', rmse)
```

```python
motor_size= 2.0
input_data = pd.DataFrame([[motor_size]], columns=['Engine Size(L)'])
predicted_co2= model.predict(input_data)
print(f'El coche con un motor de {motor_size}L produce aproximadamente {predicted_co2}gramos de CO2.')
```
